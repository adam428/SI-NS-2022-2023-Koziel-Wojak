```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.datasets import fetch_california_housing
```

```python
california_housing = fetch_california_housing(as_frame=True)
data = california_housing.data
target = california_housing.target
```

```python
# Eliminacja outlinerów

std_dev = np.std(data, axis=0)
mean = np.mean(data, axis=0)

# Określanie zakresu wartości odstających
outlier_range = 3 * std_dev

# Sprawdzanie, które wartości w danych przekraczają zakres odchylenia standardowego
outliers = np.abs(data - mean) > outlier_range

# Usuwanie wartości odstających
clean_data = data[~np.any(outliers, axis=1)]
clean_target = target[~np.any(outliers, axis=1)]
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Tworzenie instancji obiektu MinMaxScaler
min_max_scaler = MinMaxScaler()

# Przeskalowanie danych za pomocą MinMaxScaler
normalized_data = min_max_scaler.fit_transform(clean_data)

# Tworzenie instancji obiektu StandardScaler
standard_scaler = StandardScaler()

# Standaryzacja danych za pomocą StandardScaler
standardized_data = standard_scaler.fit_transform(clean_data)
```

```python
# Podział danych na zbiór treningowy i testowy
X_train, X_test, y_train, y_test = train_test_split(normalized_data, standardized_data, test_size=0.2, random_state=42)
```

```python
# Inicjalizacja modelu
model = LinearRegression()

# Trenowanie modelu na danych treningowych
model.fit(X_train, y_train)
```

```python
# Przewidywanie na danych testowych
y_pred = model.predict(X_test)

# Obliczenie błędów
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```

```python
from sklearn.preprocessing import PolynomialFeatures

# Tworzenie instancji obiektu PolynomialFeatures
poly_features = PolynomialFeatures(degree=5)

# Przekształcanie danych treningowych na cechy wielomianowe
X_train_poly = poly_features.fit_transform(X_train)

# Przekształcanie danych testowych na cechy wielomianowe
X_test_poly = poly_features.transform(X_test)
```

```python
# Tworzenie instancji modelu regresji liniowej
model_poly = LinearRegression()

# Trenowanie modelu na danych treningowych
model_poly.fit(X_train_poly, y_train)

# Przewidywanie na danych testowych
predictions_poly = model_poly.predict(X_test_poly)

# Obliczanie błędu średniokwadratowego (MSE)
mse_poly = mean_squared_error(y_test, predictions_poly)

# Obliczanie współczynnika determinacji (R2 score)
r2_poly = r2_score(y_test, predictions_poly)

# Wyświetlanie wyników
print("Błąd średniokwadratowy (MSE):", mse_poly)
print("Współczynnik determinacji (R2 score):", r2_poly)
```
